# Email Validation (string Filtering + Loops)

```python
# Assignment1
def isValid(email):
    if not email or len(email) == 0:
        return False
    
    count = 0
    for char in email:
        if char == '@':
            count += 1
    
    if count != 1:
        return False
    
    #check for '.com' or '.org'
    com = False
    org = False

    if len(email) >= 4:
        if email[-4:] == '.com':
            com = True
    
    if len(email) >= 4:
        if email[-4:] == '.org':
            org = True
    
    if com or org:
        return True
    else:
        return False

umail = input("Enter an email address to validate: ")

if isValid(umail):
    print(f"✓ '{umail}' is VALID!")
else:
    print(f"✗ '{umail}' is NOT VALID!")
```

# Count Word Frequency in Customer Reviews
Analyzing reviews or comments in text analytics

```python
# Assignment2
def countfreq(text):
    tlow = ''   #lowercase
    for char in text:
        if 'A' <= char <= 'Z':
            tlow += chr(ord(char) + 32) #uppercase to lowercase
        else:
            tlow += char
    
    words = []
    current = ''
    for char in tlow:
        if ('a' <= char <= 'z') or ('0' <= char <= '9'):
            current += char
        else:
            if current:
                words.append(current)
                current = ''
    
    if current:
        words.append(current)
    word_count = {}
    
    for word in words:
        if word in word_count:
            word_count[word] = word_count[word] + 1
        else:
            word_count[word] = 1
    return word_count

review = input("Enter a customer review: ")
frequency = countfreq(review)
print(f"Word Frequency:\n{frequency}")
```

# Password Strength Checker

```python
#Assignment3
def check(password):
    if len(password) < 8:
        return "Weak"
    
    upper = False
    lower = False
    digit = False
    special = False
    schar = "!@#$%^&*()-+"
    
    for char in password:
        if 'A' <= char <= 'Z':
            upper = True
        elif 'a' <= char <= 'z':
            lower = True
        elif '0' <= char <= '9':
            digit = True
        elif char in schar:
            special = True
    
    if upper and lower and digit and special:
        return "Strong"
    else:
        return "Moderate"

upass = input("Enter a password: ")
strength = check(upass)
print(f"The password strength is: {strength}")
```
