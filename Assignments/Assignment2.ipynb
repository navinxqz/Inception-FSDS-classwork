# Email Validation (string Filtering + Loops)

```python
# Assignment1
def isValid(email):
    if not email or len(email) == 0:
        return False
    
    count = 0
    for char in email:
        if char == '@':
            count += 1
    
    if count != 1:
        return False
    
    #check for '.com' or '.org'
    com = False
    org = False

    if len(email) >= 4:
        if email[-4:] == '.com':
            com = True
    
    if len(email) >= 4:
        if email[-4:] == '.org':
            org = True
    
    if com or org:
        return True
    else:
        return False

umail = input("Enter an email address to validate: ")

if isValid(umail):
    print(f"âœ“ '{umail}' is VALID!")
else:
    print(f"âœ— '{umail}' is NOT VALID!")
```

# Count Word Frequency in Customer Reviews
Analyzing reviews or comments in text analytics

```python
# Assignment2
def countfreq(text):
    tlow = ''   #lowercase
    for char in text:
        if 'A' <= char <= 'Z':
            tlow += chr(ord(char) + 32) #uppercase to lowercase
        else:
            tlow += char
    
    words = []
    current = ''
    for char in tlow:
        if ('a' <= char <= 'z') or ('0' <= char <= '9'):
            current += char
        else:
            if current:
                words.append(current)
                current = ''
    
    if current:
        words.append(current)
    word_count = {}
    
    for word in words:
        if word in word_count:
            word_count[word] = word_count[word] + 1
        else:
            word_count[word] = 1
    return word_count

review = input("Enter a customer review: ")
frequency = countfreq(review)
print(f"Word Frequency:\n{frequency}")
```

# Password Strength Checker

```python
#Assignment3
def check(password):
    if len(password) < 8:
        return "Weak"
    
    upper = False
    lower = False
    digit = False
    special = False
    schar = "!@#$%^&*()-+"
    
    for char in password:
        if 'A' <= char <= 'Z':
            upper = True
        elif 'a' <= char <= 'z':
            lower = True
        elif '0' <= char <= '9':
            digit = True
        elif char in schar:
            special = True
    
    if upper and lower and digit and special:
        return "Strong"
    else:
        return "Moderate"

upass = input("Enter a password: ")
strength = check(upass)
print(f"The password strength is: {strength}")
```

# Username Generator

```python
#Assignment4
def generator(fname):
    names = fname.split()
    
    if len(names) < 2:
        return "Error: Please enter both first and last name"
    
    firstName = names[0]
    lastName = names[1]
    
    first_part = ''.join([chr(ord(c) + 32) if 'A' <= c <= 'Z' else c for c in firstName[:3]])   #using lc style
    last_part = ''.join([chr(ord(c) + 32) if 'A' <= c <= 'Z' else c for c in lastName[-3:]])
    randomNum = (sum([ord(c) for c in fname]) % 90) + 10
    return first_part + last_part + str(randomNum)

fullName = input("Enter your full name (First+Last): ")
username = generator(fullName)
print(f"Generated Username: {username}")
```

# SMS Word Counter
Used in messaging apps or social media post limit check.

```python
#Assignment5
def countSMS(message):
    charCount = len(message)
    wordCount = 0
    inWord = False
    
    for char in message:
        if char != ' ' and char != '\t' and char != '\n':
            if not inWord:
                wordCount += 1
                inWord = True
        else:
            inWord = False
    if charCount <= 160:
        status = "Within Limit"
    else:
        status = "Exceeds Limit"
    return wordCount, charCount, status

message = input("Enter your message: ")
words, characters, status = countSMS(message)
print(f"Words = {words}, Characters = {characters}, Status = {status}")
```

# Remove Duplicates from Contact List
Real-world use: Data cleaning in CRM or user databases.

```python
#Assignment6
def remover(contact):
    names = [name.strip() for name in contact.split(',')]   #using lc
    unique = []
    [unique.append(name) for name in names if name and name not in unique]
    return ', '.join(unique)


contacts = input("Enter names: ")
uniqueContacts = remover(contacts)
print(f"Original: {contacts}\nAfter removing duplicates: {uniqueContacts}")
```

# Detect Suspicious Words in Text (Moderation)
Used in chat moderation or email filtering.

```python
#Assignment7
def detector(text):
    banned_words = ['spam', 'scam', 'fake']
    
    text_lower = ''
    for char in text:
        if 'A' <= char <= 'Z':
            text_lower += chr(ord(char) + 32)
        else:
            text_lower += char
    
    words = []
    current = ''
    
    for char in text_lower:
        if ('a' <= char <= 'z') or ('0' <= char <= '9'):
            current += char
        else:
            if current:
                words.append(current)
                current = ''
    
    if current:
        words.append(current)
    for word in words:
        for banned in banned_words:
            if word == banned:
                return True
    
    return False

user_text = input("Enter text: ")
if detector(user_text):
    print("Warning: Suspicious content detected!")
else:
    print("Content is safe.")
```

# Count Hashtags and Mentions (Social Media Analytics)
ðŸ’¼ Real-world use: Used in analyzing social media posts.

```python
#Assignment9
def counter(post):
    hashtagCount = sum([1 for char in post if char == '#'])
    mentionCount = sum([1 for char in post if char == '@'])
    return hashtagCount, mentionCount

post = input("Enter a social media post: ")
hashtags, mentions = counter(post)
print(f"Hashtags = {hashtags}, Mentions = {mentions}")
```

# Word Reversal for Data Encryption
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.

```python
#Assignment10
def reversal(sentence):
    reversedWords = [word[::-1] for word in sentence.split()]
    return ' '.join(reversedWords)

text = input("Enter a sentence: ")
reversedText = reversal(text)
print(f"Original: {text}\nReversed: {reversedText}")
```
