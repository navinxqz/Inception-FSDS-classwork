# Email Validation (string Filtering + Loops)

```python
# Assignment1
def isValid(email):
    if not email or len(email) == 0:
        return False
    
    count = 0
    pos = -1
    for i in range(len(email)):
        if email[i] == '@':
            count += 1
            pos = i
    
    if count != 1:
        return False
    
    local = email[:pos]
    dom = email[pos + 1:]
    
    #local part
    if not local or len(local) == 0:
        return False
    
    if local[0] == '.' or local[-1] == '.':
        return False
    
    pchar = ''
    for char in local:
        if char == '.' and pchar == '.':
            return False
        
        isValid = False
        if 'a' <= char <= 'z' or 'A' <= char <= 'Z':
            isValid = True
        elif '0' <= char <= '9':
            isValid = True
        elif char in '.\_-+':
            isValid = True
        
        if not isValid:
            return False
        
        pchar = char
    
    #domain part
    if not dom or len(dom) == 0:
        return False
    
    labels = []
    current = ''
    for char in dom:
        if char == '.':
            if current:
                labels.append(current)
                current = ''
        else:
            current+= char
    
    if current:
        labels.append(current)
    if len(labels) < 2:
        return False
    
    for i in range(len(labels)):
        label = labels[i]
        
        if not label or len(label) == 0:
            return False
        
        if label[0] == '-' or label[-1] == '-':
            return False
        
        # Check characters in label
        for char in label:
            is_valid = False
            if 'a' <= char <= 'z' or 'A' <= char <= 'Z':
                is_valid = True
            elif '0' <= char <= '9':
                is_valid = True
            elif char == '-':
                is_valid = True
            
            if not is_valid:
                return False
        
        if i == len(labels)-1:
            if len(label)<2:
                return False
            
            for char in label:
                if not ('a' <= char <= 'z' or 'A' <= char <= 'Z'):
                    return False
    
    return True

umail = input("Enter an email address to validate: ")

if isValid(umail):
    print(f"✓ '{umail}' is VALID!")
else:
    print(f"✗ '{umail}' is NOT VALID!")
```
