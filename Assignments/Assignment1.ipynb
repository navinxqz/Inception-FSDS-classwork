# Assignment 2

```python
# ATM withdraw sys
bal = float(input("Your account balance: "))
w = float(input("Amount to withdraw: "))
if w > bal:
    print("Insufficient Balance")
else:
    bal -= w
    print("Transaction Successful")
    print(f"Remaining Balance: {bal}")
```

# Assignment 3

```python
#student grade calc
marks = float(input("Input marks: "))
if 90 <= marks <= 100:
    grade = 'A+'
elif 80 <= marks < 90:
    grade = 'A'
elif 70 <= marks < 80:
    grade = 'B'
elif 60 <= marks < 70:
    grade = 'C'
else:
    grade = 'Fail'
print(f"Your grade is: {grade}")
```
