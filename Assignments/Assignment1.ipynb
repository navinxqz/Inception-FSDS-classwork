# Assignment 2

```python
# ATM withdraw sys
bal = float(input("Your account balance: "))
w = float(input("Amount to withdraw: "))
if w > bal:
    print("Insufficient Balance")
else:
    bal -= w
    print("Transaction Successful")
    print(f"Remaining Balance: {bal}")
```

# Assignment 3

```python
#student grade calc
marks = float(input("Input marks: "))
if 90 <= marks <= 100:
    grade = 'A+'
elif 80 <= marks < 90:
    grade = 'A'
elif 70 <= marks < 80:
    grade = 'B'
elif 60 <= marks < 70:
    grade = 'C'
else:
    grade = 'Fail'
print(f"Your grade is: {grade}")
```

# Assignment 4

```python
# bus ticket discount
ticket = 500
age = int(input("Input your age: "))

if age < 5:
    discount = 100 
elif 5 <= age <= 18:
    discount = 50
elif age >= 60:
    discount = 30
else:
    discount = 0

d = (discount/100)*ticket
price = ticket-d

print(f"Your bus ticket discount is: {discount}%")
# print(f"Discount amount: {d} Tk")
print(f"Your final ticket price is: {price} Tk")
```

```python
#restaurant bill with discount
bill = float(input("Your total bill amount: "))
if bill > 1000:
    discount = 10
    d = (discount/100)*bill
    final = bill-d

    print(f"{discount}% discount applied.")
    print(f"Your final bill amount is: {final} Tk")
else:
    print("No discount applied.")
    print(f"Your final bill amount is: {bill} Tk")
```

# Assignment 8

```python
#leap year checker
year = int(input("Input a year: "))

if year % 400 == 0:
    print(f"{year} is a leap year.")
elif (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
