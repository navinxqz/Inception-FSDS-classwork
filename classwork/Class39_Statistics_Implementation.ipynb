## Measure of central Tendency
- Mean
- Median
- Mode

```python
ages = [25, 30, 35, 40, 45, 50, 55, 60, 35, 37, 22, 28, 32, 38, 42, 48, 32, 38, 26,100]
```

```python
import numpy as np
```

```python
print(np.mean(ages))
```

```python
print(np.median(ages))  #no impact of outlier
```

```python
import statistics as st
```

```python
print(st.median(ages))
```

```python
print(st.mode(ages))    #numpy doesn't have mode func
```

```python
!pip install seaborn
import seaborn as sns
```

### 5 number summary

```python
q1,q3=np.percentile(ages,[25,75])
```

```python
print(q1,q3)
```

```python
IQR=q3-q1
lower_fence=q1-1.5*IQR
upper_fence=q3+1.5*IQR
print(lower_fence, upper_fence)
```

## Measure of dispersion
- Variance
- Standard Deviation

```python
st.variance(ages)
```

```python
st.stdev(ages)
```

```python
st.pvariance(ages)
```

```python
import math
math.sqrt(st.pvariance(ages)) #population standard deviation
```

### Other distribution
- Lognormal distribution
- Power law distribution

```python
#lognormal distribution
mu,sigma=3.,1. #mean and standard deviation
s=np.random.lognormal(mu,sigma,1000)
sns.histplot(s, kde=True)
```

```python
#lognormal distribution to normal distribution
sns.histplot(np.log(s), kde=True)
```

```python
sns.histplot(np.log(ages), kde=True)
```

```python
!pip install scipy
import scipy.stats as stats
import pylab as pl
import matplotlib.pyplot as plt
```

```python
def plot_data(sample):
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(sample, kde=True)
    plt.title('Data Distribution')
    plt.subplot(1, 2, 2)
    stats.probplot(sample, dist="norm", plot=pl)
    plt.title('Probability Plot')   #QQ plot to check normality of data
    plt.show()
```

```python
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
plot_data(ages)
```
