regualr expression or "regex" for short, are a powerful tool for working with strings and text data in python. They allow us to match and manipulate strings based on pattern, making it easy to perform complex string operatios with just a few lines of code

```python
t = text.split(' ')

for word in t:
    if word == "Cyclone":
        print("Found")
```

```python
import re
```

```python
pattern = r"[A-Z]+yclone"
match=re.search(pattern, text)
print(match)
```

```python
match_item=re.finditer(pattern, text)
for match in match_item:
    index=match.span()
    print(f"Match found at index: {index} -> {text[index[0]:index[1]]}")
```
