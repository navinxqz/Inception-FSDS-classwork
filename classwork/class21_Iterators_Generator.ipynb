## Iterator
### what is an Iteration
Iteration is a general term for taking each item of something one after another. Any time we use a loop, explicit or implicit, to go over a group of items, that is iteration.

```python
num = [1,2,3]
for i in num:
    print(i)
```

### what is Iterator
An Iterator is an obj that allows the programmer to traverse through a seq of data without having to store the entire data in the memory.

```python
L = [i for i in range(1,10)]
# List is an iterable obj so we can perform loop/ iteration on it 
L
```

```python
import sys
print(sys.getsizeof(L))
#  iterable obj take memomry once
# iterable obj can crush the ROM memory when working with big data
# that is why iterators are used to save memory
```

```python
x = range(1,100000)
print(sys.getsizeof(x))
# range is also an iterable obj but it takes less memory than list
# takes the memomry one by one when iterated
```

### making our own for loop

```python
def my_for_loop(iterable):
    iterator = iter(iterable)
    while True:
        try:
            element = next(iterator)
            print(element)
        except StopIteration:
            break
```

```python
a=[10,20,30]
b=range(1,10)
c=(1,2,3)
d={1,2,3}
e={0:1,1:2,2:3}
```

```python
my_for_loop(a)
```

```python
my_for_loop(e)
```

```python
my_for_loop(b)
```

## Generator

```python
def gen():
    yield 1 #if we see yield keyword then it is generator function
    yield "statement"
    yield "67"
    #return give the output and exit the function, but yield give the output and pause the function
```

```python
gen=gen()
```

```python
gen
```

```python
#to get the values from generator we use iteration -> next() function
for i in gen:
    print(i)
```
