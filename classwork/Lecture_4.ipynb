# Accessing Strings

```python
st="hello world"
print(st[0])  # Accessing first character

len(st)  # Length of the string
```

```python
#negative indexing
print(st[-1])  # Accessing last character
print(st[-5])  # Accessing fifth character from the end
```

```python
#string slicing
print(st[0:5])  # Slicing from index 0 to 4
print(st[6:])   # Slicing from index 6 to end
print(st[:5])  # Slicing from start to index 4
print(st[:])   # Slicing the entire string
print(st[::2])  # Slicing with step 2 (Jump parameters)
print(st[1::2])  # Slicing with step 2 starting from index 1
print(st[::-1])  # Reversing the string
print(st[-1:-6:-1])  # Slicing in reverse from last character to fifth character from the end
print(st[-5:])  # Slicing last five characters
print(st[-6:-1])  # Slicing from sixth last to second last character
print(st[0:7:2])  # Slicing from index 0 to 6 with step 2
```

### Python strings are immutable

```python
#arithmetic operations on strings
s1 = "Hello"
print(s1 + " World")  # Concatenation
print(s1 * 3)  # Repeating the string 3 times
```

```python
print("|" * 10)
```

```python
#relational operations on strings

print("apple" == "apple")  # True
print("apple" != "banana")  # True
print("apple" < "banana")  # True
print("banana" > "apple")  # True
print("apple" <= "apple")  # True
print("banana" >= "apple")  # True
print("apple" < "Apple")  # False, because lowercase letters have higher ASCII values than uppercase letters
print("Apple" < "apple")  # True
print("apple" < "apricot")  # True, comparison is done lexicographically
print("apricot" > "apple")  # True
print("apple" < "applepie")  # True, shorter string is considered smaller if it is a prefix of the longer string
print("applepie" > "apple")  # True
```

```python
#membership operations on strings
'N' in "Navin"
'A' not in 'Amrin'
```

# Common Function/Method
- len

```python
s = "hello world"
len(s)
```

```python
max(s) # 'w' has the highest ASCII value
```

```python
min(s) # ' ' (space) has the lowest ASCII value
```

```python
sorted(s)  # Returns a list of sorted characters in the string
```

```python
sorted(s, reverse=True)  # Returns a list of sorted characters in descending order
```

```python
s.capitalize()  # Capitalizes the first character of the string
```

```python
s.title()  # Capitalizes the first character of each word in the string
```

```python
s.upper()  # Converts all characters to uppercase
```

```python
s = 'Welcome'
s.swapcase()  # Swaps the case of each character
```

```python
s ="My name is Navin"
s.count('n')  # Counts occurrences of 'n' (case-sensitive)
```

```python
s.find('n')  # Finds the first occurrence of 'n' and returns its index
```

```python
s.endswith('d')  # Checks if the string ends with 'd'
```

```python
s.startswith('My')  # Checks if the string starts with 'h'
```
