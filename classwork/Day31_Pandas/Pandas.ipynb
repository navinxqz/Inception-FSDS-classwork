### Adding/Removing Columns

```python
coffee.head()
```

```python
coffee["price"]=4.99
```

```python
coffee.head()
```

```python
import numpy as np
```

```python
coffee["new_price"] = np.where(coffee["Coffee Type"] == "Espresso",3.99,5.99)
```

```python
coffee.head(7)
```

```python
coffee.drop(columns=["price"], inplace=True) # drop the price column
# inplace=True will modify the original dataframe, if inplace=False it will return a new dataframe with the column dropped but the original dataframe will remain unchanged
# for permanent delete use inline=True
```

```python
coffee.head()
```

```python
coffee["revenue"]=coffee["Units Sold"]*coffee["new_price"]
```

```python
coffee.head()
```

```python
# rename
coffee.rename(columns={"new_price":"price"}, inplace=True)
```

```python
def categorized_athlete(row):
    if row["height_cm"]<175 and row["weight_kg"]<70:
        return "Lightweight"
    elif row["height_cm"]<185 or row["weight_kg"]<=80:
        return "Middleweight"
    else:
        return "Heavyweight"
```

```python
bios["Category"] = bios.apply(categorized_athlete, axis=1)
```
