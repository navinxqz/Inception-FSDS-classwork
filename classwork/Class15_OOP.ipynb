```python
class Robot:
    def __init__(self, robot_num):
        self.robot_num = robot_num

    def movement(self): #method
        print(f"Robot number is {self.robot_num}: moving forward.")

    def fireDetection(self):
        print(f"Robot number is {self.robot_num}: fire detected.")

    def wheelControl(self):
        print(f"Robot number is {self.robot_num}: wheels are moving.")
```

```python
robot1 = Robot(1)
```

```python
robot1.wheelControl()
```

```python
robot2 = Robot(2)
```

```python
robot2.movement()
```
