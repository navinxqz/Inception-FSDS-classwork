# Characteristics of a List
- Ordered
- changeable/mutable
- can be duplicates
- can be nested
- items can be accessed
- Dynamic Size
- Hetrogeneous
- Speed of execution -> slow
- more memory
- multiple data type store

```python
L=[1,2,3]
```

```python
print(id(L))
print(id(L[0]))
```

# create a list

```python
print([])   # empty list
print([1,2,3])  # 1D
print([1,2,3,[4,5,6]])  # 2D
print([1,2,3,[4,5,[6,7]]])  # 3D

print([1,True,5.6,6+7j,"Hello"])  # Hetrogeneous List
print(list("Hello"))  # converting other data types to list
```

# Accessing Items from a List
- indexing
- slicing

```python
l=[1,2,3,4,5]
print(l[2])
print(l[-3])
```

```python
l2=[1,2,3,[4,5]]
print(l2[3])
print(l2[3][1])
```

```python
print(l[:])
print(l[1:4])
print(l[::2])
print(l[::-1])
```

# List's Methods

```python
l.append(6)
print(l)
```

```python
#editing with slicing
L=[10,20,30,40,50]
L[1:4]=[200,300,400]
print(L)
```

```python
L=[10,20,30,40,50]
L.insert(2,25)
print(L)
```

```python
del L[1:3]
print(L)
```

```python
L.remove(10)
print(L)
```
