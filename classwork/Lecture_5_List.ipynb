# Operations of lists
- Arithmetic
- Membership
- Loop

```python
L1 = [1,2,3]
L2 = [4,5,6]
```

```python
L1 + L2 # concatenation
```

```python
L1 * 3  # repetition
```

```python
#membership
L1 = [1,2,3,4,5]
L2 = [4,5,[6,7,8]]
```

```python
print(3 in L1)
print(40 in L2)
print(7 in L2[2])
print([6,7,8] not in L2)
```

```python
#list comprehension
LC = [i for i in range(1,11)]
print(LC)
```

```python
#scalar multiplication on vectors
v = [1,2,3,4,5]
s = -3

L=[]
for i in v:
    L.append(s*i)
print(L)
```

```python
LC = [i*s for i in v]
print(LC)
```

```python
#print all numbers from 1 to 50 which are divisible by 5
LC = [i for i in range(1,51) if i%5==0]
print(LC)
```

```python
#find lang which start with 'P' from the given list
langs = ['Python', 'Java', 'C', 'C++', 'Perl', 'PHP', 'JavaScript']
PC_langs = [lang for lang in langs if lang.startswith('P')]
print(PC_langs)
```

# Zip
The zip func returns the zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterator have different lengths, the least items decides the length of the new iterator.

```python
L1=[1,2,3]
L2=['a','b','c']
```

```python
#zip(L1,L2)
list(zip(L1,L2))
```

```python
roll_no = [101,102,103,104]
names = ['Alice', 'Bob', 'Charlie', 'David']
list(zip(roll_no,names))
```

```python
L1=[1,2,3,4,5]
L2=[6,7,8,9,10]

[i-j for i,j in zip(L2,L1)]
```

```python
L1=[1,2,3,4,5]
L2=[4,5,6,7,8]
L3=[9,10,11]

list(zip(L1,L2,L3))
```
