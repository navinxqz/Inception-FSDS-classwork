There are 2 stages where error may happen:
- Durning compilation -> Syntax Error
- During execution -> Exceptions

### Syntax Error
- Something in the program is not written according to the program grammar
- Error is raised by the interpreter/complier
- can solve by rectifying the program

```python
# example of syntax error
print "Hello, World!"
```

### Exceptions
If things go wrong during the execution of the program (runtime). It generally happens when something unforeseen has happened.
- Exceptions are raised by python runtime
- have to tackle on the fly

#### Examples
- Memory overflow
- Divide by -> logical error
- Database error

```python
def divide(a):
    r = 10 / a
    print(r)
```

```python
divide(5)
```

```python
divide(0)
```

```python
try:
    with open("files/file.txt", "r") as f:
        content = f.read()
        print(content)
except:
    print("File not found!")
```

```python
def divide_num(a,b):
    if b==0:
        raise Exception("Denominator cannot be zero.")
    else:
        return a/b
```

```python
divide_num(10, 0)
```
