# Loops in python

```python
#while loop
i=0
while i<10:
    print("Iteration:", i)
    i+=1
```

```python
#multiplication table using while loop
num = int(input("Enter a number: "))
i=1
while i<=10:
    print(f"{num} x {i} = {num*i}")
    i+=1
```

# Guessing Game

```python
import random
jackpot = random.randint(1, 100)
# print(jackpot)
guess = int(input("Guess the jackpot number between 1 and 100: "))
counter = 1

while guess != jackpot:
    if guess < jackpot:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")
    guess = int(input("Enter your new guess: "))
    counter += 1
print(f"Congratulations! You've guessed the jackpot number in {counter} attempts!")
```

```python
for i in range(1, 6):
    print("Iteration:", i)
```

```python
# nested loops
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i: {i}, j: {j}")
```

# Triangle prints

```python
for i in range(1, 6):
    for j in range(1, i+1):
        print("*", end=" ")
    print()
```

```python
n=int(input("Enter the number: "))

for i in range(1, n+1):
    for j in range(1, i+1):
        print("*", end=" ")
    print()
```

# Loop control statements
- Break
- Continue
- Pass

```python
#break
for i in range(1, 11):
    if i == 6:
        break
    print(i)    
```

```python
#continue
for i in range(1, 11):
    if i == 5:
        continue
    print(i)    
```

```python
#pass
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```
