ATM machine - ATMMachine

```python
class AtmMachine:
    #constructor (special func) --> super power
    def __init__(self):
        # print("ATM Machine is ready to use")
        self.balance = 0
        self.pin = ""
        self.menu()
    
    def menu(self):
        user_input = input("""
        Welcome to ATM Machine!
        Please choose your activity:
        1 --> Create Pin
        2 --> Change Pin
        3 --> Check Balance
        4 --> Withdraw
        5 --> Exit
        """)
        
        if user_input == "1":
            self.createPin()
        elif user_input == "2":
            self.changePin()
        elif user_input == "3":
            self.checkBalance()
        elif user_input == "4":
            self.withdraw()
        elif user_input == "5":
            print("Thank you for using the ATM Machine. Goodbye!")
            exit()
        else:
            print("Invalid input. Please try again.")
            self.menu()
    
    def createPin(self):
        userPin = input("Please enter your new PIN: ")
        self.pin = userPin

        userBalance = float(input("Please enter your initial balance: "))
        self.balance = userBalance
        print("PIN created successfully!")
        self.menu()

    def changePin(self):
        oldPin = input("Please enter your old PIN: ")
        if oldPin == self.pin:
            newPin = input("Please enter your new PIN: ")
            self.pin = newPin
            print("PIN changed successfully!")
        else:
            print("Incorrect PIN. Please try again.")
        self.menu()
    
    def checkBalance(self):
        userPin = input("Please enter your PIN: ")
        if userPin == self.pin:
            print("Your current balance is: ",self.balance)
        else:
            print("Incorrect PIN. Please try again.")
        self.menu()
    
    def withdraw(self):
        userPin = input("Please enter your PIN: ")
        if userPin == self.pin:
            amount = float(input("Please enter the amount to withdraw: "))
            if amount <= self.balance:
                self.balance -= amount
                print("Withdrawal successful! Your new balance is: ",self.balance)
            else:
                print("Insufficient balance.")
        else:
            print("Incorrect PIN. Please try again.")
        self.menu()
```
