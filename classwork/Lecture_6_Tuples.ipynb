# Tuples
similar to a list. But we can't change the element of a tuples once it is assigned. Tuple is an immutable list.

### Characteristics:
- ordered
- Unchangeable
- Allow duplicate

```python
# Creating tuple
t1 = ()
print(t1)
print(type(t1))
```

```python
t2 =("Navin",)  # single element tuple (,) must
print(t2)
print(type(t2))

t3 =("Navin")
print(t3)
print(type(t3))
```

```python
t4 = (1, 2, 3, 4, 5)    #hetro
print(t4)
print(type(t4))

t5 = (1, "Navin", 3.4, True, [1, 2, 3])  #hetro
print(t5)

t6 = tuple('Hello') # converting string to tuple
print(t6)

```

# Type Casting

```python
t3 = (1, 2, 3, 4, 5)
temp = list(t3)  # converting tuple to list
print(temp)
```

```python
temp.append(6)
print(temp)
```

```python
#tuple take less memory than list
import sys
L= list(range(1000))
T= tuple(range(1000))

print(sys.getsizeof(L))  # size of list in bytes
print(sys.getsizeof(T))  # size of tuple in bytes
```

```python
#Zipping tuples
a=(1,2,3,4)
b=('a','b','c','d')
tuple(zip(a,b))
```

```python
#scalar multiplication on a vector
v=[1,2,3]
s=-3
L=[]

for i in v:
    result= s*i
    L.append(result)
print(L)
```

```python
TC=tuple((i*s) for i in v) #tuple comprehension
print(TC)
```
