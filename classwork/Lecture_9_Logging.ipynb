```python
import logging
import os
```

```python
LOG_DIR = "logs"
LOG_FILE_NAME = "app.log"

os.makedirs(LOG_DIR, exist_ok=True)
log_path=os.path.join(LOG_DIR, LOG_FILE_NAME)
print(log_path)

logging.basicConfig(
    filename=log_path,
    format="[%(asctime)s] %(name)s - %(levelname)s - %(message)s",
    level=logging.INFO
)
```

```python
def profile(name):
    try:
        if type(name) == str:
            print(f"welcome {name}")
    except Exception as e:
        logging.info("Invalid name format. Name must be a string.")
```

```python
profile('1234')
```

```python
def divide_num(a,b):
    if b==0:
        logging.info("Can't divide num with zero.")
    else:
        return a/b
```

```python
divide_num(3, 3)
```
